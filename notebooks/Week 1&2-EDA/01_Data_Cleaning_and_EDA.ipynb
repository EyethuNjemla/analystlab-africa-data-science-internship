# Data Quality Assessment

Before cleaning the datasets, it is important to assess their quality.

Data quality issues such as missing values, duplicate records, incorrect data types, and inconsistent values can negatively affect analysis and machine learning models.

In this section, the following checks will be performed for both datasets:

- Missing values
- Duplicate records
- Data types
- Number of unique values

```python
# FUNCTION TO ASSESS DATA QUALITY
#Generate a summary of the dataset's quality.
# Returns:
#        Summary table showing data types,
#        missing values, percentage missing,
#        duplicate count and unique values.
def data_quality(df):

    # Creating a summary DataFrame
    quality = pd.DataFrame({
        "Data Type": df.dtypes,
        "Missing Values": df.isnull().sum(),
        "Percentage Missing (%)": round((df.isnull().sum()/len(df))*100,2),
        "Unique Values": df.nunique()
    })

    return quality
```

```python
# Displaying the data quality for the titanic dataset
print("TITANIC DATA QUALITY REPORT")
display(data_quality(titanic))
```

```python
# CHECKING FOR DUPLICATE RECORDS

# Counting duplicate rows in the Titanic dataset
titanic_duplicates = titanic.duplicated().sum()

# Counting duplicate rows in the Housing dataset
housing_duplicates = housing.duplicated().sum()

# Display the results
print(f"Titanic duplicate records : {titanic_duplicates}")
print(f"Housing duplicate records : {housing_duplicates}")
```

```python
# CHECKING FOR MISSING VALUES

# Calculating missing values for each column in the Titanic dataset
titanic_missing = titanic.isnull().sum().sort_values(ascending=False)

# Calculating missing values for each column in the Housing dataset
housing_missing = housing.isnull().sum().sort_values(ascending=False)

print("MISSING VALUES - TITANIC DATASET")
display(titanic_missing)

print("MISSING VALUES - HOUSING DATASET")

display(housing_missing)
```

# Data Cleaning

The objective of this section is to improve the quality of the datasets by addressing common data issues.

The following tasks will be performed:

- Remove duplicate records
- Handle missing values
- Verify data types
- Confirm that the datasets are ready for exploratory data analysis

```python
# REMOVING DUPLICATE RECORDS

# Remove duplicate rows from the Titanic dataset
titanic.drop_duplicates(inplace=True)

# Remove duplicate rows from the Housing dataset
housing.drop_duplicates(inplace=True)

# Display the updated number of duplicate rows
print("Duplicate Records After Cleaning")
print(f"Titanic : {titanic.duplicated().sum()}")
print(f"Housing : {housing.duplicated().sum()}")
```

```python
# HANDLE MISSING VALUES - TITANIC DATASET

try:
    print("Trying to handle missing values in the titanic dataset")

    print("Missing Values Before Cleaning")
    display(titanic.isnull().sum())
    # Filling missing Age values using the median
    titanic["Age"] = titanic["Age"].fillna(titanic["Age"].median())

    # Filling missing Embarked values using the most frequent value
    titanic["Embarked"] = titanic["Embarked"].fillna(titanic["Embarked"].mode()[0])

    # Dropping the Cabin column because it contains too many missing values
    if "Cabin" in titanic.columns:
        titanic = titanic.drop(columns="Cabin")

    print("Titanic missing values handled successfully.")

except Exception as ex:
    print(f"An error occured: {ex}")
```

```python
 # HANDLE MISSING VALUES - HOUSING DATASET
try:
    print("Handling missing values in the Housing dataset")

    # Displaying missing values before cleaning
    print("Missing Values Before Cleaning")
    display(housing.isnull().sum())

    # Identify numerical columns
    numerical_columns = housing.select_dtypes(include=np.number).columns

    # Filling missing numerical values with the median
    for column in numerical_columns:
        housing[column] = housing[column].fillna(housing[column].median())

    # Identifing categorical columns
    categorical_columns = housing.select_dtypes(exclude=np.number).columns

    # Filling missing categorical values with the mode
    for column in categorical_columns:
        housing[column] = housing[column].fillna(housing[column].mode()[0])

    print("Housing missing values handled successfully.")

except Exception as e:
    print(f"An error occured: {e}")
```

```python
# VERIFYING THAT ALL MISSING VALUES HAVE BEEN HANDLED
# Verifying for the titanic dataset
print("Remaining Missing Values - Titanic")
display(titanic.isnull().sum())

# verifying for the housing dataset
print("Remaining Missing Values - Housing")
display(housing.isnull().sum())
```

```python
# FUNCTION TO SAVE THE CLEANED DATASETS AS CSV FILES

def save_clean_dataset(df, filename):
   
    # Define the output directory
    output_path = "../../datasets/processed/"

    # Create the full file path
    full_path = output_path + filename

    # Save the DataFrame as a CSV file without the index
    df.to_csv(full_path, index=False)

    # Display a confirmation message
    print(f"Dataset successfully saved to:\n{full_path}")
```

```python
# Saving both the cleaned titanic and housing datasets
try:
    print("Saving the cleaned datasets as csv files.")
    # Saving the cleaned Titanic dataset
    save_clean_dataset(titanic, "Titanic_Cleaned.csv")
    # Saving the cleaned Housing dataset
    save_clean_dataset(housing, "Housing_Cleaned.csv")

    print("Datasets saved succesfully!")

except Exception as e:
    print(f"The following error occured:{e}")
```

# Data visualistions

```python
# titatnic survival count
plt.figure(figsize=(7,5))
sns.countplot(data=titanic, x="Survived")
plt.title("Passenger Survival Count")
plt.xlabel("Survived")
plt.ylabel("Number of Passengers")
plt.savefig(
    "../../assets/plots/titanic_survival_count.png",
    dpi=300,
    bbox_inches="tight"
)
plt.show()
```

```python
# titatnic survival by Gender
plt.figure(figsize=(7,5))
sns.countplot(data=titanic, x="Sex", hue="Survived")
plt.title("Passenger Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")
plt.savefig(
    "../../assets/plots/titanic_survival_by_gender.png",
    dpi=300,
    bbox_inches="tight"
)
plt.show()
```

```python
# NUMBER OF SURVIVORS BY PASSENGER CLASS

# Count the number of survivors in each passenger class
survival_by_class = (
    titanic[titanic["Survived"] == 1]["Pclass"]
    .value_counts()
    .sort_index()
)

# Create the bar chart
plt.figure(figsize=(8, 5))

sns.barplot(
    x=survival_by_class.index.astype(str),
    y=survival_by_class.values
)

# Add chart title and axis labels
plt.title("Number of Survivors by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Survivors")

# Add value labels on top of each bar
for index, value in enumerate(survival_by_class.values):
    plt.text(index, value + 2, str(value), ha="center")

plt.savefig(
    "../../assets/plots/survivors_by_class.png",
    dpi=300,
    bbox_inches="tight"
)

# Display the chart
plt.show()
```

```python
# SURVIVAL DISTRIBUTION BY PASSENGER CLASS

# Filter the dataset to include only passengers who survived
survivors = titanic[titanic["Survived"] == 1]

# Count the number of survivors in each passenger class
survival_by_class = survivors["Pclass"].value_counts().sort_index()

# Create a pie chart
plt.figure(figsize=(8, 8))

plt.pie(
    survival_by_class,
    labels=[f"Class {cls}" for cls in survival_by_class.index],
    autopct="%1.1f%%",
    startangle=90,
    explode=(0.03, 0.03, 0.03)
)

# Add a title
plt.title("Distribution of Survivors by Passenger Class")

# Ensure the pie chart is circular
plt.axis("equal")

plt.savefig(
    "../../assets/plots/titanic_survival_PieChart.png",
    dpi=300,
    bbox_inches="tight"
)

# Display the chart
plt.show()
```

```python
# AVERAGE HOUSE PRICE BY NUMBER OF BEDROOMS

# Calculate the average price for each bedroom category
average_price = housing.groupby("bedrooms")["price"].mean()

# Create the figure
plt.figure(figsize=(10, 6))

# Plot the average prices
sns.barplot(
    x=average_price.index,
    y=average_price.values
)

# Add a title
plt.title("Average House Price by Number of Bedrooms")

# Label the axes
plt.xlabel("Number of Bedrooms")
plt.ylabel("Average House Price")

# Add value labels above each bar
for index, value in enumerate(average_price.values):
    plt.text(
        index,
        value,
        f"{value:,.0f}",
        ha="center",
        va="bottom",
        fontsize=9
    )

# Save the figure 
plt.savefig(
    "../../assets/plots/Ave_price_bedrooms.png",
    dpi=300,
    bbox_inches="tight"
)
# Display the plot
plt.show()
```
